# Accessing scores 

```python
from astropy.table import Table
scores = Table.read('path/to/outlier_scores_alt.txt', format="ascii")
scores.to_pandas(index='KIC')
```

## Example of organizing data to see the most outlying objects in Quarter 16

```python
scores_df = scores.to_pandas(index='KIC')
```

```python
ranks_by_Q16 = scores_df.rank(ascending=False).sort_values(by='Q16')
ranks_by_Q16
```
